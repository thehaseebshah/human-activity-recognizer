##Now, XGBoost

```python
import xgboost as xgb
```

```python
Dtrain = xgb.DMatrix(X_train, y_train)
```

```python
Dtrain
```

```python
 Dtest = xgb.DMatrix(X_test, y_test)
```

```python
steps = 20
```

```python
 model = xgb.train(param, Dtrain, steps)
```

```python
 import numpy as np
```

```python
 
from sklearn.metrics import accuracy_score
```

```python
preds = model.predict(Dtest)
```

```python
preds.shape
```

```python
 best_preds = np.asarray([np.argmax(line) for line in preds])
```

```python
best_preds
```

```python
best_preds.shape
```

```python
 print(f"Accuracy = {accuracy_score(y_test, best_preds)}")
```

##Trying Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
 gbm_param_grid = {
    'colsample_bytree': [0.3, 0.7],
    'n_estimators': [50],
    'max_depth': [2, 5]
}
```

```python
 gbm = xgb.XGBRegressor()
# Perform grid search: grid_mse
grid_rmse = GridSearchCV(estimator=gbm, param_grid=gbm_param_grid,
scoring='neg_mean_squared_error', cv=3, verbose=1)
```

```python
grid_rmse.fit(X,y)
```

```python
xg_reg = grid_rmse.best_estimator_
```

```python
 print("Best parameters found: ", grid_rmse.best_params_)
print("Lowest RMSE found: ", np.sqrt(np.abs(grid_rmse.best_score_)))
```
